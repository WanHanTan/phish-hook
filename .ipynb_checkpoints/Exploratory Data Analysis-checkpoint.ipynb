```python
from wordcloud import WordCloud

# Text Processing
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
import string
from nltk.tokenize import word_tokenize
```

```python
malicious_url = urldata[urldata['label'] == 'malicious']['url']
benign_url = urldata[urldata['label'] == 'benign']['url']

malicious_words = []
benign_words = []

def extractMaliciousWords(maliciousUrl):
    global malicious_words
    words = [word.lower() for word in word_tokenize(maliciousUrl) if word.lower() not in stopwords.words("english") and word.lower().isalpha()]
    malicious_words = malicious_words + words
    
def extractBenignWords(benignUrl):
    global benign_words
    words = [word.lower() for word in word_tokenize(benignUrl) if word.lower() not in stopwords.words("english") and word.lower().isalpha()]
    benign_words = benign_words + words

malicious_url.apply(extractMaliciousWords)
benign_url.apply(extractBenignWords)
```

```python
malicious_wordcloud = WordCloud(background_color = 'black', width=600, height=400, collocations =False).generate(" ".join(malicious_words))
plt.figure(figsize=(10,8),facecolor = 'k',edgecolor = 'k')
plt.imshow(malicious_wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
                            
```

```python
def visualize(label):
    words = ''
    for i in urldata[urldata['label'] == label]['url']:
        i = i.lower()
        words += i + ''
    wordcloud = WordCloud(width=600, height=400, collocations=False).generate(words)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

```python
visualize('malicious')
```

```python
visualize('benign')
```
